## Prepare proposals dataset

```python
sample = 100
```

```python
models = list(set(val.columns) - set(["name", "description"]))
grouped_proposals = [list(x) for x in zip(*[val[model][:sample] for model in models])]
```

```python
from itertools import chain
import numpy as np
from ranker import add_dcg_weights
from torch.utils.data import DataLoader 

def encode_proposals(proposals, names):
    X = []
    y = []
    for props, name in zip(proposals, names):
        unique_proposals = list(set(chain(*props)))
        if name not in unique_proposals:
            continue
        models_weights = {}
        for model, proposals_model in zip(models, props):
            weighted = add_dcg_weights(proposals_model)
            models_weights[model] = {name: w for w, name in weighted}
        pattern = [
            [
                models_weights[model][name] if name in models_weights[model] else 0
                for model in models
            ]
            for name in unique_proposals
        ]
        X.append(pattern)
        y.append(unique_proposals.index(name))
    return X, y
```

```python
X, y = encode_proposals(grouped_proposals, val["name"][:sample])
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SelfAttentionRanker(nn.Module):
    def __init__(self, n_models, n_heads=3, hidden_size=2048, dropout=0.1, n_layers=2):
        super(SelfAttentionRanker, self).__init__()

        self.n_models = n_models
        
        transformer_layer = nn.TransformerEncoderLayer(n_models, n_heads, dim_feedforward=hidden_size, dropout=dropout)
        self.mixer = nn.TransformerEncoder(transformer_layer, num_layers=n_layers)
        self.collapser = nn.Linear(n_models, 1, bias=False)

    def forward(self, x):
        x = self.mixer(x)
        x = self.collapser(x)
        return F.softmax(x, dim=1)
```
