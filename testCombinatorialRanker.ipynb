## Combinatorial ranker

Try all possible combinations of models, in all possible orders

```python
from collections import Counter
from model import dcg

def majority_ranking(proposals):
    return [[elem[0] for elem in Counter(names).most_common(10)] for names in proposals]

def combine_models(dataset, included_models):
    # Join columns of included models, in the order proposed
    proposals = [[] for _ in range(len(dataset))]
    for model in included_models:
        proposals = [p + added for p, added in zip(proposals, dataset[model])]
    return majority_ranking(proposals)

def combination_quality(dataset, included_models):    
    return dcg(combine_models(dataset, included_models), dataset["name"])
```

```python
from itertools import permutations
import numpy as np
from tqdm import tqdm_notebook

def combinatorial_ranker(dataset):
    models = list(set(val.columns) - set(["name", "description"]))
    metrics = {}
    # Try all ensemble sizes
    for n_models in tqdm_notebook(range(1, len(models) + 1)):
        for included_models in tqdm_notebook(permutations(models, n_models)):
            metrics[tuple(included_models)] = combination_quality(dataset, included_models)
    # Recover winning combination
    best_models = list(metrics.keys())[np.argmax(list(metrics.values()))]
    combination = combine_models(dataset, best_models)
            
    return combination, best_models, metrics
```

```python
%%time
combination, best_models, metrics = combinatorial_ranker(val)
metrics
```
