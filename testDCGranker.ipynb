## Test ranking on validation data

```python
sample = len(val)
```

```python
def combine_proposals(dataset):
    models = list(set(val.columns) - set(["name", "description"]))
    proposals = [[] for _ in range(len(dataset))]
    for model in models:
        proposals = [p + added for p, added in zip(proposals, dataset[model])]
    return proposals
```

```python
from ranker import majority_ranking

majority_ranked_proposals = majority_ranking(combine_proposals(val.iloc[:sample]))
```

```python
majority_ranked_proposals
```

```python
from ranker import dcg_ranking

models = list(set(val.columns) - set(["name", "description"]))
grouped_proposals = [list(x) for x in zip(*[val[model][:sample] for model in models])]
dcg_ranked_proposals = dcg_ranking(grouped_proposals)
```

```python
dcg_ranked_proposals
```

```python
from model import dcg

print(f"DCG majority ranked   = {dcg(majority_ranked_proposals, val['name'][:sample])}")
print(f"DCG DGC-weight ranked = {dcg(dcg_ranked_proposals, val['name'][:sample])}")
```
